```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def exp(x, y):
    return (np.exp(x)*np.cos(y), np.exp(x)*np.sin(y))
```

```python
x = np.linspace(start=-3, stop=20, num=100)
y = np.linspace(start=-3, stop=20, num=100)
```

```python
X, Y = np.meshgrid(x, y)
Z_re = np.array([[exp(_x, _y)[0] for _x, _y in zip(_X, _Y)] for _X, _Y in zip(X, Y)]) # 実部
Z_im = np.array([[exp(_x, _y)[1] for _x, _y in zip(_X, _Y)] for _X, _Y in zip(X, Y)]) # 虚部
```

```python
%matplotlib notebook
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X,Y,Z_re, color="red", label="関数値の実部")
ax.plot_wireframe(X,Y,Z_im, color="blue", label="関数値の虚部")
plt.legend()
ax.set_xlabel("引数の実部")
ax.set_ylabel("引数の虚部")
ax.set_zlabel("関数値")
```
